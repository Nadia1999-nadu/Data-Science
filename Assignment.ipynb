```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv')
df
```

```python
df.isnull()
```

```python
# Numbers of Relatives
df['Numbers of relatives'] = df['SibSp'] + df['Parch']
df
```

```python
#Scatter plt of the passengers who were alive
survived_df = df[df['Survived'] == 1]
survived_df
```

```python
# Scatter
plt.figure(figsize=(10, 8))
plt.scatter(survived_df['Numbers of relatives'], survived_df['Age'], alpha=0.9, color='#420')
plt.xlabel('Number of Relatives')
plt.ylabel('Age')
plt.title('Scatter Plot of Survived Passengers')
plt.grid()
plt.show()
```

```python
# bar
plt.figure(figsize=(8,6))
survived = survived_df.groupby('Numbers of relatives')['Age'].mean()
survived_df.groupby('Numbers of relatives')['Age'].mean().plot(kind = 'bar',alpha = 0.7 ,color='red')
plt.xlabel('Number of Relatives')
plt.ylabel('Average Age')
plt.title('Bar Plot of Average Age by Number of Relatives for Survived Passengers')
plt.show()
```
