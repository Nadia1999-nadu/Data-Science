```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df1 = pd.read_csv('term-test-1-result.csv')
df1
```

```python
df2 = pd.read_csv('term-test-2-result.csv')
df2
```

```python
#df2 = pd.read_csv('term-test-2-result.csv')
#df2
```

```python
# Merge dataframes on 'Registration Number'
merged_df = pd.merge(df1, df2, on=['Registration Number','Name'], suffixes=('TT-1 Marks', 'TT-2 Marks'))

print("Files merged successfully!")
```

```python
merged_df
```

```python
merged_df['Average TT Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)
```

```python
merged_df
```

```python
merged_df['Best TT Marks'] = merged_df[['TT-1 Marks' , 'TT-2 Marks']].max(axis = 1)
```

```python
merged_df
```

```python
merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'], inplace=True)
```

```python
merged_df
```

```python
attendance_df = pd.read_csv('attendance-term_final.csv')
```

```python
attendance_df
```

```python
final_merged = pd.merge(merged_df, attendance_df, on=['Registration Number','Name'], suffixes=('_TT', '_Attendance'))
```

```python
final_merged
```

```python
# Merge with latest term test dataframe
#final_df = pd.merge(merged_df, attendance_df, on=['Registration Number','Name'], suffixes=('_TT', '_Attendance'))
```

```python
#final_df
```

```python
# Compute final marks
final_merged['Final Marks'] = (final_merged['Term full marks'] * 0.7) + final_merged['Average TT Marks'] + final_merged['Attendence']
```

```python
final_merged
```

```python
# Save to final result CSV
final_merged.to_csv('final_result.csv', index=False)
```

```python
final_merged
```

```python
# K-Means Clustering
X = final_merged[['Final Marks']]
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    final_merged[f'Cluster_{k}'] = kmeans.fit_predict(X)
```

```python
# Visualize the clusters for k=3 (example)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
final_merged['Cluster'] = kmeans.fit_predict(X)
plt.scatter(final_merged['Registration Number'], final_merged['Final Marks'], c=final_merged['Cluster'], cmap='viridis')
plt.xlabel('Registration Number')
plt.ylabel('Final Marks')
plt.title('K-Means Clustering of Final Marks')
plt.colorbar(label='Cluster')
plt.show()
```
